**Prueba de Kolmogorov - Smirnov**

Donde debemos calcular $D^+$, $D^-$ y $D$.

Donde: $D^+=max_{1<i<n}|{\frac{i}{n}-x_{i}|}$, $D^-=max_{1<i<n}|{x_{i}-\frac{i-1}{n}|}$ y $D = max(D^+,D^-)$.

El valor critico $d_{\alpha,n}$.

Y se acepta si $D <d_{\alpha,n}$



Nivel de aceptación de 95 %.

```python
from tabulate import tabulate
lista = [0.78, 0.98, 0.24, 0.73, 0.43, 0.16, 0.78, 0.47, 0.18, 0.55,
          0.04, 0.29, 0.68, 0.77, 0.16, 0.03, 0.79, 0.22, 0.37, 0.80,
          0.96, 0.26, 0.91, 0.55, 0.75, 0.55, 0.64, 0.39, 0.53, 0.45,
          0.61, 0.14, 0.38, 0.12, 0.40, 0.74, 0.78, 0.98, 0.27, 0.60,
          0.43, 0.67, 0.62, 0.32, 0.53, 0.54, 0.24, 0.29, 0.18, 0.08,
          0.82, 0.94, 0.19, 0.98, 0.41, 1.00, 0.74, 0.92, 0.14, 0.43,
          0.83, 0.88, 0.18, 0.21, 0.50, 0.13, 0.43, 0.69, 0.08, 0.12,
          0.22, 0.50, 0.16, 0.11, 0.18, 0.89, 0.80, 0.42, 0.29, 0.87,
          0.83, 0.79, 0.65, 0.28, 0.78, 0.49, 0.36, 0.86, 0.87, 0.64,
          0.51, 0.07, 0.18, 0.94, 0.50, 0.22, 0.66, 0.91, 0.48 ,0.24]
lista_ordenada = sorted(lista) #Esta funcion nos devuelve una nueva lsita
tabla = []
Darriba = []
Dabajo = []
n = len(lista_ordenada)
i= 1

for elemento in lista_ordenada:
  Dmas = abs(((i-1)/n)-elemento)
  Darriba.append(Dmas)
  Dmenos = abs((elemento)-((i-1)-1)/n)
  Dabajo.append(Dmenos)
  iteracion = [i,elemento, Dmas, Dmenos]
  tabla.append(iteracion)
  i+=1

H_0="Los datos estan en una distribucion"
H_a = "Los datos no estan distribucion"
punto = round(0.875897/(n**0.5),4)
mas = max(Darriba)
menos = max(Dabajo)
ambas = max(mas, menos)
print(f"D+={mas}, D-={menos}, D=max(D+,D-): {ambas}, d={punto} y {n}")
if ambas < punto:
  print(f"Se acepta H_0: {ambas}<{punto}")
else:
  print(H_a)

print(tabulate(tabla,headers = ("i","Xi", "D+", "D-"),  tablefmt='fancy_grid'))

```
