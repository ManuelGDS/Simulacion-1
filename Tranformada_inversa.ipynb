# **Método de la transformada inversa**

See $f(x)=\frac{x^2}{9}$ si $0\leq x\leq 3$

Comprobamos que $f(x)$ es una FDP

1) $f(x)\geq 0$ en $[0,3]$

2) $\int_{-\infty}^{\infty}f(x)dx=\frac{x^3}{27}|_{0}^{3}=(\frac{27}{27})-(\frac{0}{27})=1$

Cumple ambas propiedades $\therefore es una FDP$

$F(x)=\int_{-\infty}^{\infty}f(s)ds=\int_{0}^{x}\frac{s^2}{9}ds=\frac{x^3}{27}$

Entonces

$F(x)=u$

$u=\frac{x^3}{27}->x = 3 \cdot\sqrt[3]{u}$

```python
import numpy as np
import random as rnd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,3,100)
y=x**2/9

plt.figure(figsize=(8,6))
plt.plot(x,y)
```

```python
def inversa(n):
  lista = []
  for i in range(n):
    u = rnd.random()
    x = 3*(u)**(1/3)
    lista.append(x)
  return lista
```

```python
G = inversa(10000)
```

```python
plt.hist(G, bins=50)
```

```python
def cdf(G): #Función de distribución acumulada
 n=len(G)
 X=np.sort(G)
 Y=np.arange(0,n)/n

 return X,Y
```

```python
X, Y=cdf(G)
```
