Algoritmo para la generación de números pseudoaleatorios pormedio del **"Metodo de Congruencia"** y esta dado por la forma
$x_{n+1} = ax_n + c \pmod{m},  0<x_n<m$

Donde: $x_0$ es la semilla, $a$, $c$ y  $m$ son constantes.


(a) xj+1 = (21xj + 15)mod(31) con x0 = 21

```python
from tabulate import tabulate
x_0 = 21 #int(input("Ingrese la semilla (X_0): "))
a = 21 #int(input("Ingrese la constante a: "))
c = 15 #int(input("Ingrese la constante c: "))
m =31 #int(input("Ingrese el modulo: "))
n = 50 #int(input("Ingrese el numero de interaciones a desarrolar: "))


matriz = []

for i in range(0, n+1):

  xi = (a*x_0)+ c
  xmod = xi % (m)
  xn = x_0/m
  xiter = [i, x_0, xi, xmod, xn]
  matriz.append(xiter)
  x_0 = xmod

print(tabulate(matriz,headers = ("N","Xn", "aXn+C", "aXn + C (mod)", "Xn/mod" ),  tablefmt='grid'))
```
